```python
from gtts import gTTS
from weather import Weather
import pyttsx3
import speech_recognition as sr
import os
import sys
import webbrowser
import smtplib
import requests
import datetime
import random
import wikipedia
import wolframalpha
import re

engine = pyttsx3.init()
#engine.say("I will speak this text")
engine.runAndWait()
client = wolframalpha.Client('xxxxxxxxxxxxx') # your wolframa app ID
def speak(audio):
    print('Computer: ' + audio)
    engine.say(audio)
    engine.runAndWait()
    

def greetMe():
    currentH = int(datetime.datetime.now().hour)
    if currentH >= 0 and currentH < 12:
        speak('Good Morning!')

    if currentH >= 12 and currentH < 18:
        speak('Good Afternoon!')

    if currentH >= 18 and currentH !=0:
        speak('Good Evening!')

greetMe()

speak('Hello Sir, I am your digital assistant!')
speak('How may I help you?')

def myCommand():
    "listens for commands"

    r = sr.Recognizer()

    with sr.Microphone() as source:
        print('Ready...')
        r.pause_threshold = 1
        r.adjust_for_ambient_noise(source, duration=1)
        audio = r.listen(source)

    try:
        command = r.recognize_google(audio).lower()
        print('You said: ' + command + '\n')

    #loop back to continue to listen for commands if unrecognizable speech is received
    except sr.UnknownValueError:
        speak('Sorry sir! I didn\'t get that! Try typing the command!')
        #print('Your last command couldn\'t be heard')
        command = myCommand();

    return command


def assistant(command):
    
    "if statements for executing commands"
    
    if 'open youtube' in command:
            speak('okay')
            webbrowser.open('www.youtube.com')

    elif 'open google' in command:
            speak('okay')
            webbrowser.open('www.google.co.in')

    elif 'open gmail' in command:
            speak('okay')
            webbrowser.open('www.gmail.com')

    elif "what\'s up" in command or 'how are you' in command:
            stMsgs = ['Just doing my thing!', 'I am fine!', 'Nice!', 'I am nice and full of energy']
            speak(random.choice(stMsgs))

    elif 'email' in command:
            speak('Who is the recipient? ')
            recipient = myCommand()

            if 'me' in recipient:
                try:
                    speak('What should I say? ')
                    content = myCommand()
        
                    server = smtplib.SMTP('smtp.gmail.com', 587)
                    server.ehlo()
                    server.starttls()
                    server.login("mail id ", 'password')
                    server.sendmail('mail id', "Recipient_Username", content)
                    server.close()
                    speak('Email sent!')

                except:
                    speak('Sorry Sir! I am unable to send your message at this moment!')
                    
    
    elif 'current weather' in command or 'weather forecast' in command:
        
            speak('Can you please tell me the city name ? ')
            city = myCommand()            
    
            def weather_data(query):
                res=requests.get('http://api.openweathermap.org/data/2.5/weather?'
                                 +query+'&APPID=xxxxxxxxxxxxxxxxxxxxxx &units=metric');                                   
                return res.json();
            def print_weather(result,city):
                speak("{}'s temperature: {}Â°C ".format(city,result['main']['temp']))
                speak("Wind speed: {} m/s".format(result['wind']['speed']))
                speak("Description: {}".format(result['weather'][0]['description']))
                speak("Weather: {}".format(result['weather'][0]['main']))
            def main():
                #reg_ex = re.search('current weather in (.*)', command)
                #city=input('Enter the city:')
                #city=reg_ex
                print()
                try:
                    query='q='+city;
                    w_data=weather_data(query)
                    print_weather(w_data, city)
                    print()
                except:
                    speak('City name not found...')
            if __name__=='__main__':
                main()
                
    elif 'joke' in command:
        res = requests.get(
                'https://icanhazdadjoke.com/',
                headers={"Accept":"application/json"}
                )
        if res.status_code == requests.codes.ok:
            speak(str(res.json()['joke']))
            speak("ha..ha..ha..ha..ha..")
        else:
            speak('oops!I ran out of jokes')


    elif 'nothing' in command or 'abort' in command or 'stop' in command:
            speak('okay')
            speak('Bye Sir, have a good day.')
            sys.exit()
           
    elif 'hello' in command:
            speak('Hello Sir')

    elif 'bye' in command:
            speak('Bye Sir, have a good day.')
            sys.exit()
                                    
    elif 'play music' in command:
            music_folder = "start D:/music/"
            music = ['m','m1','m2','m3']
            print(random.choice(music))
            random_music = music_folder + random.choice(music) + '.mp3'
            os.system(random_music)
                  
            speak('Okay, here is your music! Enjoy!')
            

    else:
            command = command
            speak('Searching...')
            try:
                try:
                    res = client.command(command)
                    results = next(res.results).text
                    speak('WOLFRAM-ALPHA says - ')
                    speak('Got it.')
                    speak(results)
                    
                except:
                    results = wikipedia.summary(command, sentences=2)
                    speak('Got it.')
                    speak('WIKIPEDIA says - ')
                    speak(results)
        
            except:
                webbrowser.open('www.google.com')
        
    speak('Next Command! Sir!')

    



#loop to continue executing multiple commands
while True:
    assistant(myCommand())

```
